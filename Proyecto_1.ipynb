```python
with open("datos.txt", "w") as archivo:
    archivo.write("Fanny\n")
    archivo.write("Nubia\n")
    archivo.write("Nelson\n")
    archivo.write("Yurley\n")
    archivo.write("Huber\n")
    archivo.write("Jose\n")

print("Archivo creado")
```

```python
archivo = open("datos.txt", "r")

nombres = archivo.readlines()

archivo.close()

salidas = []

for nombre in nombres:

    mensaje = f"Hola {nombre.strip()}, bienvenido.\n"

    salidas.append(mensaje)

nuevo_archivo = open("saludos.txt", "w")

nuevo_archivo.writelines(salidas)

nuevo_archivo.close()

print("Archivo generado correctamente")
```
